```python
# ..... importing the neccessary libraries

import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data_file = pd.read_csv('iris.csv' , index_col = 'idx')
```

```python
data_file.head()
```

### Line Plot .... sns.lineplot()

```python
# Dealing with Line Plot  

# if you want to plot all columns of Dataset

plt.figure(figsize = (8,4))                      # to set the figure size
plt.title('Plotting values of Features')


# Line plot 

sns.lineplot(data = data_file)
plt.ylabel('Value')
```

```python
# if you want to plot specific columns of Dataset

plt.figure(figsize = (8,4))             # to set the figure size
plt.title('Plotting values of determinated Features')


sns.lineplot(data = data_file['sepal.length'] , label = 'sepal length')
sns.lineplot(data = data_file['sepal.width'] , label = 'sepal width')

plt.ylabel('Value')
```

### bar Plot .... sns.barplot()

```python
# Dealing with Bar Plot  

plt.figure(figsize = (8,4))             # to set the figure size
plt.title('Plotting values of determined one Feature ')

# Bar Chart
sns.barplot(x = data_file.index , y = data_file['petal.length'])

# note here x .index , because i want index to be in x 
# it will be error if you write column name 
# you can do that if it is not column index
```

### HeatMap Plot .... sns.heatmap()

```python
# Dealing with HeatMop Plot  

data_noTarget = pd.read_csv('iris_notarget.csv' , index_col = 'idx')
# i made a copy of the iris file , removing target string to be able to use Heat Map
plt.figure(figsize=(8,6))
plt.title("Heat Map of Features")

# Heat Map
sns.heatmap(data = data_noTarget , annot = True)  # you must write (annot)
```

### Scatter Plot .... sns.scatterplot()

```python
# Dealing with HeatMop Plot  


plt.figure(figsize=(8,6))
plt.title("Scatter Plot of two Features colored by other feature (categorical)")

# Scatter Plot
sns.scatterplot(x = data_file['sepal.length'] , y = data_file['sepal.width'])

# as you shown that all of points are the same color , look at the next cell 
```

```python

# working on a the scatter plot 

plt.figure(figsize=(8,6))
plt.title("Scatter Plot of two Features colored by other feature (categorical)")

# Scatter Plot
sns.scatterplot(x = data_file['sepal.length'] , y = data_file['sepal.width'] , hue = data_file['variety'])

# (hue) >> provides diffrent colors according to what you say after it 
# here i made colors by target values [type of the flower]
# what you make by orange.exe you can make it by code and it is fantastic


plt.legend()
```

```python

# working on a the scatter plot with one regression line

plt.figure(figsize=(8,6))
plt.title("Scatter Plot of two Features with one regression line")

# regplot Plot
sns.regplot(x = data_file['sepal.length'] , y = data_file['sepal.width'] )
```

```python

# working on a the scatter plot with more than one regression line

# lmplot Plot
sns.lmplot(x = 'sepal.length' , y = 'sepal.width'  , hue = 'variety', data = data_file )

# lmplot is diffrent in writing its prop.
```

### Distibutions Plot .... sns.distplot() 

```python
# Dealing on the Distributions plot 

# histogram , x = sepal length , y = distribution
sns.distplot(a = data_file['petal.length'] , kde = False)

# kde == kernel density estimate
```

```python

# working on the Distributions plot 

# kde plot in one dimension   
# you can print one or more in the same plot
sns.kdeplot(data = data_file['sepal.length'], label = 'sepal length' , shade = True)
sns.kdeplot(data = data_file['sepal.width'] , label = 'sepal width',shade = True)
sns.kdeplot(data = data_file['petal.length'] , label = 'petal length',shade = True)
sns.kdeplot(data = data_file['petal.width'] , label = 'petal width',shade = True)


plt.xlabel('')
plt.legend()
# kde == kernel density estimate
```

```python

# working on the Distributions plot 

# kde plot in two dimension (contour)
# you can print one or more in the same plot
sns.jointplot(x = data_file['sepal.width'] , y = data_file['petal.width'] , shade = True, kind = 'kde')


# kde == kernel density estimate
```
